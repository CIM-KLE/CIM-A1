```python
import csv

def parse_csv(file_path):
    results = {}
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            print(f"Row: {row}")  # Debug print to see the content of each row
            # Adjusting to the correct column names
            results['precision'] = float(row['   metrics/precision'])
            results['recall'] = float(row['      metrics/recall'])
            results['mAP@0.5'] = float(row['     metrics/mAP_0.5'])
            results['mAP@0.5:0.95'] = float(row['metrics/mAP_0.5:0.95'])
    return results

def print_metrics(results):
    print(f"Precision: {results['precision']:.2f}")
    print(f"Recall: {results['recall']:.2f}")
    print(f"mAP@0.5: {results['mAP@0.5']:.2f}")
    print(f"mAP@0.5:0.95: {results['mAP@0.5:0.95']:.2f}")

# Replace with the path to your CSV file
csv_file_path = '/content/YoloV5results.csv'

# Parse and print the results
results = parse_csv(csv_file_path)
print_metrics(results)
```

```python
import csv

def parse_csv(file_path):
    results = {}
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            print(f"Row: {row}")  # Debug print to see the content of each row
            # Adjusting to the correct column names
            results['precision'] = float(row['   metrics/precision(B)'])
            results['recall'] = float(row['      metrics/recall(B)'])
            results['mAP@0.5'] = float(row['       metrics/mAP50(B)'])
            results['mAP@0.5:0.95'] = float(row['    metrics/mAP50-95(B)'])
    return results

def print_metrics(results):
    print(f"Precision: {results['precision']:.2f}")
    print(f"Recall: {results['recall']:.2f}")
    print(f"mAP@0.5: {results['mAP@0.5']:.2f}")
    print(f"mAP@0.5:0.95: {results['mAP@0.5:0.95']:.2f}")

# Replace with the path to your CSV file
csv_file_path = '/content/yolov8results.csv'

# Parse and print the results
results = parse_csv(csv_file_path)
print_metrics(results)
```
